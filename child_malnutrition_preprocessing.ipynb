## provided data 

source: https://apps.who.int/gho/data/node.main.ngest?lang=en

```python
df_overweight = pd.read_csv("C://Users/Admin/Downloads/NUTOVERWEIGHTPREV_NUTOVERWEIGHTNUM.csv")
df_underweight = pd.read_csv("C://Users/Admin/Downloads/NUTUNDERWEIGHTPRE_NUTUNDERWEIGHTNUM.csv")
df_stunted = pd.read_csv("C://Users/Admin/Downloads/NUTSTUNTINGPREV_NUTSTUNTINGNUM.csv")
df_wasted = pd.read_csv("C://Users/Admin/Downloads/NUTWASTINGPREV_NUTWASTINGNUM.csv")
df_sevwasted = pd.read_csv("C://Users/Admin/Downloads/NUTSEVWASTINGPREV_NUTSEVWASTINGNUM.csv")
```

```python
def split_data_col(df, iloc_to_split):
    x = df.iloc[:, iloc_to_split].str.split(' ', expand = True)
    x.rename(columns = {0: 'expected value', 1 :'split_1'}, inplace=True)

    x.loc[:, 'split_1'] = x.loc[:, 'split_1'].str.strip('[]')
    temp_split = x['split_1'].str.split('-', expand = True)
    temp_split.rename( columns = {0: 'lower_CI', 1:'upper_CI'}, inplace=True)

    x = pd.concat([x, temp_split], axis=1)#.drop(1, axis=1)
    x = x.drop('split_1', axis=1)
    return x
```

```python
def split_concat(data_input, condition): 
    temp_data_2 = split_data_col(data_input, 2)
    temp_data_2.rename(columns = {'expected value': f'{condition}_expected_value_prob', 
                                  'lower_CI': f'{condition}_lower_CI_prob', 'upper_CI': f'{condition}_upper_CI_prob'}, inplace=True)
    temp_data_3 = split_data_col(data_input, 3)
    temp_data_3.rename(columns = {'expected value': f'{condition}_expected_value_millions', 
                                  'lower_CI': f'{condition}_lower_CI_millions', 'upper_CI': f'{condition}_upper_CI_millions'}, inplace=True)

    x = pd.concat([data_input, temp_data_2, temp_data_3], axis=1)
    x = x.drop(x.iloc[:, [2,3]], axis = 1)
    return x   
```

```python
# df_overweight = split_concat(df_overweight)
df_overweight = split_concat(df_overweight, 'overweight')
df_underweight = split_concat(df_underweight, 'underweight')
df_stunted = split_concat(df_stunted, 'stunted')
df_wasted = split_concat(df_wasted, 'wasted')
df_sevwasted = split_concat(df_sevwasted, 'sevwasted')
```
